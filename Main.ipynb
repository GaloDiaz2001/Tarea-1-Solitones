```python
#SOLUCIONES ESTÁTICAS
#la solución estatica
#Comprobación
#def RK4(f,y0,x0,xf,steps):
def V14int(x,y):
    return  1/(np.sqrt((1/2)*(x**2-1)**2))
yval,xval= ode.RK4(V14int,-2.5,-0.95,0.95,500)
x_x = np.linspace(-2.5,2.5,500)
yanal = np.tanh(1/(np.sqrt(2))*x_x)
plt.title("Campo sombrero mexicano")
plt.plot(xval,yval,'ro',label='numérica')
plt.plot(x_x,yanal,'-',color = 'green',label='analítica')
plt.grid()
plt.legend()
plt.savefig("CampoSombrero")
plt.show()
#La densidad de energía
#Comprobación
h = 0.01
xrho = np.arange(-2.5,2.5+h,h)
intrho = np.tanh(1/(np.sqrt(2))*xrho)
rho = 1/2*(dif.right1(intrho,h))**2
r_anal = 1/4*1/(np.cosh(x_x/(np.sqrt(2))))**4
def rhoa(x,y):
    return 1/2*1/(np.cosh(x/(np.sqrt(2)))**4)
plt.title("Densidad de energía sombrero mexicano")
plt.plot(xrho,rho,'ro',label='numérica')
plt.plot(x_x,r_anal,'-',color = 'green',label='analítica')
plt.grid()
plt.legend()
plt.savefig("EnergíaSombrero")
plt.show()
#La masa
Ex,Ey= ode.RK4(rhoa,0,-10,10,600)
Etot = Ey[-1]
plt.plot(Ex,Ey,'-',color = 'green',label='Energia')
plt.grid()
print(f"La energía total = {Etot}")
```

```python
#Ahora hacemos sine-gordon
def VSGint(x,y):
    return 1/(2*np.sin((1/2)*x))
x_x = np.linspace(-2.5,2.5,500)
yanal = 4*np.arctan(np.e**x_x)
yval,xval= ode.RK4(VSGint,-2.5,0.33,6,500)
x_x = np.linspace(-2.5,2.5,500)
plt.title("Campo Sine-Gordon")
plt.plot(xval,yval,'ro',label='numérica')
plt.plot(x_x,yanal,'-',color = 'green',label='analítica')
plt.grid()
plt.legend()
plt.savefig("CampoSineGordon")
plt.show()
#La densidad de energía
#Comprobación
h = 0.01
xrho = np.arange(-2.5,2.5+h,h)
intrho = 4*np.arctan(np.e**xrho)
rho = 1/2*(dif.right1(intrho,h))**2
r_anal = (8*np.e**(2*x_x))/((1 + np.e**(2*x_x))**2)
def rhoa(x,y):
    return (8*np.e**(2*x))/((1 + np.e**(2*x))**2)
plt.title("Densidad de energía Sine-Gordon")
plt.plot(xrho,rho,'ro',label='numérica')
plt.plot(x_x,r_anal,'-',color = 'green',label='analítica')
plt.grid()
plt.legend()
plt.savefig("EnergíaSineGordon")
plt.show()
#La masa
Ex,Ey= ode.RK4(rhoa,0,-10,10,600)
Etot = Ey[-1]
plt.plot(Ex,Ey,'-',color = 'green',label='Energia')
plt.grid()
print(f"La energía total = {Etot}")
```
